```python
import pandas as pd


df = pd.read_csv('dataforFINDS.csv')

print(df)

X = df.drop('Buy_Product', axis=1) 
y = df['Buy_Product']


print("\nFeatures (X):")
print(X)

print("\nTarget (y):")
print(y)
```

```python
print(X.shape[0]) #Row
print(X.shape[1]) #Column
```

```python
for i in range(X.shape[0]):
    if y.iloc[i] == "Yes":  
        X.iloc[i,0] = "Old"
```

```python

List = [None] * X.shape[1]

for i in range(X.shape[0]):  
    if y.iloc[i] == "Yes":  

        if all(x is None for x in List):
            List = X.iloc[i].tolist() 
        else:
            for j in range(X.shape[1]):
                if List[j] != X.iloc[i, j]:  
                    List[j] = '?'

print(List)
```
