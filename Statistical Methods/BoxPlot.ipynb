# BOX PLOT

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


df = pd.read_csv('StudentMarks.csv')


print(df)
```

```python
marks_column = df['Marks']
print(marks_column)
```

```python
marks_array  = np.array(marks_column)
```

```python
fig = plt.figure(figsize =(10, 7))

# Create the boxplot
box = plt.boxplot(marks_array, 
                  vert=True,          
                  patch_artist=True,  
                  boxprops=dict(facecolor="skyblue", color="blue"),
                  whiskerprops=dict(color="green", linewidth=2),
                  capprops=dict(color="red", linewidth=2),
                  flierprops=dict(marker='o', color='purple', markersize=8))

plt.legend([box["boxes"][0]], ['Marks Distribution'], loc='upper right')

plt.title("Boxplot of Marks")
plt.xlabel("Marks")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

# Pie Chart

```python
#No. of students in each sport
data = {"Football" : 53, "Cricket" : 42, "Badminton" : 32, "BasketBall": 12, "Table Tennis":21}
```

```python
data
```

```python
# Pie chart
plt.figure(figsize=(10, 10))
plt.pie(data.values(), labels=data.keys(), autopct='%1.1f%%')

# Title
plt.title('Number of Students in Each Sport')

plt.legend(data.keys(), title="Sports")
plt.show()
```
